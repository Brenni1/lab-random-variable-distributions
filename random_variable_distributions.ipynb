```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt



pd.set_option('display.max_rows', 50)
pd.set_option('display.max_columns', 50)
```

```python
customer_df = pd.read_csv('./files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv')
customer_df.columns = [i.lower().replace(' ', '_') for i in customer_df.columns]
customer_df.head()  #starting with the fresh file, as the old notebook just got too bloated
```

```python
data_num = customer_df.select_dtypes(np.number)
data_num
```

```python
data_num.describe()
```

```python
for column in data_num:
    sns.distplot(data_num[column])
    plt.show()

```

```python
import scipy.stats as stats
```
